```python
new_survey = {'Not interested':[], 'Somewhat interested':[], 'Very interested':[]}
```

```python
list(survey.columns)
```

```python
survey.drop(['Timestamp'], axis=1, inplace=True)
```

```python
new_survey = {'Not interested':[], 'Somewhat interested':[], 'Very interested':[], 'Index':[]}
```

```python
def make_dataframe(data):
    new_survey = {'Not interested':[], 'Somewhat interested':[], 'Very interested':[], 'Index':[]}
    for name in list(data.columns):
        temp = data[name].value_counts()
        new_survey['Very interested'].append(temp.loc['Very interested'])
        new_survey['Somewhat interested'].append(temp.loc['Somewhat interested'])
        new_survey['Not interested'].append(temp.loc['Not interested'])
        new_survey['Index'].append(temp.name.split('[')[1][:-1])
    return(pd.DataFrame.from_dict(new_survey))
```

```python
tmp = make_dataframe(survey)
```

```python
for name in list(survey.columns):
    temp = survey[name].value_counts()
    new_survey['Very interested'].append(temp.loc['Very interested'])
    new_survey['Somewhat interested'].append(temp.loc['Somewhat interested'])
    new_survey['Not interested'].append(temp.loc['Not interested'])
    new_survey['Index'].append(temp.name.split('[')[1][:-1])
```

```python
survey_clean = pd.DataFrame.from_dict(new_survey)
```

```python
survey_clean.set_index('Index', inplace=True)
```

```python
survey_clean
```

```python
survey_clean.sort_index()
```

```python
survey_clean = survey_clean[["Very interested", "Somewhat interested","Not interested"]]
```

```python
survey_clean
```

```python
survey_clean.sort_values(by="Very interested", ascending=False, inplace=True)
```

```python
survey_clean
```

```python
def percentage(a,i):
    return(a/i)
    
survey_clean.apply(func=percentage, axis=1, args=(,))
```

```python
survey_clean.sum(axis=1)
```

```python
i=0
survey_percentage = pd.DataFrame()
for sum_of_row in survey_clean.sum(axis=1):
    row = survey_clean.iloc[i, :].apply(lambda x:round((x/sum_of_row)*100,2))
    row = row.to_frame().T
    survey_percentage = pd.concat([survey_percentage,row], axis=0)
    i+=1
```

```python
survey_percentage
```

```python
%matplotlib inline

import matplotlib as mpl
mpl.style.use('ggplot') # optional: for ggplot-like style
```

### Matplot barchart

```python
%matplotlib inline

groups = 6
#plt.figure(figsize=(20,8))

#plt.figure(figsize=(20, 8))
#plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

fig, ax = plt.subplots(figsize=(20,10))
index = np.arange(groups)

bar_width = 0.2

opacity = 1
error_config = {'ecolor': '0.3'}
 

rects1 = ax.bar(index,survey_percentage["Very interested"] , bar_width,
                alpha=opacity, color="#5cb85c",
                label="Very interested")

rects2 = ax.bar(index + bar_width, survey_percentage["Somewhat interested"], bar_width,
                alpha=opacity, color="#5bc0de",
                label="Somewhat interested")

rects3 = ax.bar(index + bar_width*2, survey_percentage["Not interested"], bar_width,
                alpha=opacity, color="#d9534f",
                label="Not interested")

ax.set_title('Percentage of Respondents Interest in Data Science Areas', fontdict={'fontsize':16})
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(survey_percentage.index, rotation="vertical", fontdict={'fontsize':14})

def auto_lable(rects):
    for rect in rects:
        height = rect.get_height()
        #print(height)
        ax.text(rect.get_x() + rect.get_width()/2., 1.01*height,
                "{0:.2f}".format(height),
                ha='center', va='bottom', fontdict={'fontsize':12})
auto_lable(rects1)
auto_lable(rects2)
auto_lable(rects3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(False)
ax.get_yaxis().set_ticks([])
ax.legend()

plt.show()
```
