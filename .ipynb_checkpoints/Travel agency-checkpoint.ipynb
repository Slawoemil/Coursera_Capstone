# Travel agency

### Problem description:  
Let's say we are travel agancy and we would like to organize a trip to eight wholly different cities over 100K citiziens in USA.
Difference is defined based on dissimilarities in numbers of any kind of people's ventures in uptown part of cities.
People don't want to feel like: 'I've seen it before' and want to give a chance to smaller cities thanks to randome proceses. It's a 'unnetflix' way. 'I don't want to watch similar movies, I'm fed up, I want to try something new' people thinking like that are main target of the agency. 

### Data to be used:  
Wikipedia data to get all the cities in USA that have population over 100K citizens, along with coordinates
https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population
Foursquare data to gather information about all the venues in radius 20000 meteres around the city center.
Data will be split into eight clusters, and based on that 100 combinations(possible different trips) will be provided for the CEO of agency.  

### Methodology
Since I'll be looking for dissimilar cities I can use clustering to group similar cities into eight clusters and then choose randomly one city form each cluster to arrange whole trip with eight whereabouts.
Having these places I will add starting point which will be a somebody's home supposedly.
Finally I will render a map with trip points, and popup description showing ten most frequently occurring venues.

```python
def get_nearby_venues(city, latitudes, longitudes, radius=500, limit=10):
    i = 0
    venues_list=[]
    try:
        for city, lat, lng in zip(city, latitudes, longitudes):
            i+=1   
            # create the API request URL
            url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
                CLIENT_ID, 
                CLIENT_SECRET, 
                VERSION, 
                lat, 
                lng, 
                radius, 
                limit)   
            # make the GET request
            results = requests.get(url).json()['response']['groups'][0]['items']
            # return only relevant information for each nearby venue
            venues_list.append([(
                city,
                lat, 
                lng, 
                v['venue']['name'], 
                v['venue']['location']['lat'], 
                v['venue']['location']['lng'],  
                v['venue']['categories'][0]['name']) for v in results])

        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        nearby_venues.columns = ['City', 
                      'City Latitude', 
                      'City Longitude', 
                      'Venue', 
                      'Venue Latitude', 
                      'Venue Longitude', 
                      'Venue Category']
    except:
        None
    print('Number of added cities: ', i)
    return(nearby_venues)
```

#### Preparing trips  
Based on clusters there can be prapared many different trips all over the county.
I need additional function to add starting point excluding those trips that have the same coordinates like starting point.

```python
def prepare_n_trips(cities, number_of_trips=5):
    '''cities a dataframe with cities to visit plus clusters
    n number of trips
    '''
    number_of_clusters = len(cities['Cluster'].unique())
    trips = []
    for n in range(number_of_trips):
        one_trip = pd.DataFrame()
        for c in range(0,number_of_clusters):
            city = pd.DataFrame.sample(cities[cities['Cluster'] == c])
            one_trip = one_trip.append(city)
        trips.append(one_trip.drop('Cluster', axis=1))
    return trips    
```

```python
def add_starting_point(start, trip, first=True):
    start = pd.DataFrame({'City':start[0],
                        'Lat':start[1],
                        'Lng':start[2]}, index=[0])
    if first:
        for city in trip:
            if not start.isin(city).all().all():
                return pd.concat([start, city], axis=0, ignore_index=True)
    index = np.random.choice(len(trip))
    return pd.concat([start, trip[index]], axis=0, ignore_index=True)
```

#### Visulizing one trip with popups describing each place

```python
def visulize_trip(start_location, zoom_start, one_trip):
    trip_map = folium.Map(location=[start_location[0], start_location[1]], zoom_start=4)
    index_first = 0
    index_last = one_trip.shape[0] - 1 
    trip_ = one_trip.get_values()
    most_popular = cities_and_venues_sum.T
    for trip in trip_:
        try:
            one_city = most_popular[trip[0]]
            one_city_sorted = one_city.sort_values(ascending=False)[:10]
            desc = ''.join(('<b>',trip[0], '</b><br />'))
            for name, value in zip(one_city_sorted.index.tolist(), one_city_sorted.values.tolist()):
                desc = ''.join((desc, name, ': ', str(value), '<br />'))
        except:
            desc = '<b>Home</b>'
        if index_first == 0:
            folium.Marker(location=[trip[1], trip[2]], icon=folium.Icon(color='green', icon='home'),
                         popup=folium.Popup('{}'.format(desc))).add_to(trip_map)
        elif index_first == index_last:
            folium.Marker(location=[trip[1], trip[2]], icon=folium.Icon(color='red'),
                     popup=folium.Popup('{}'.format(desc))).add_to(trip_map)
        else:
            folium.Marker(location=[trip[1], trip[2]], icon=folium.Icon(color='blue'),
                     popup=folium.Popup('{}'.format(desc))).add_to(trip_map)
        index_first += 1
    for trip_start, trip_stop in zip(trip_[:-1], trip_[1:]):
        folium.PolyLine(locations=[trip_start[1:3], trip_stop[1:3]], color='blue').add_to(trip_map)

    all_arrows = []
    for trip_start, trip_stop in zip(trip_[:-1], trip_[1:]):
        arrows = get_arrows(locations=[trip_start[1:3], trip_stop[1:3]], n_arrows=1)
        all_arrows.append(arrows)

    for arrows in all_arrows:
        for arrow in arrows:
            arrow.add_to(trip_map)
    return trip_map
```

### Arrows  

To make milestones of a trip visible, in a ordered maner, I've borrowed two functions from Mr. Bob Haffner to create arrows.    
source: https://medium.com/@bobhaffner/folium-lines-with-arrows-25a0fe88e4e

```python
from collections import namedtuple
```

```python
def get_arrows(locations, color='blue', size=6, n_arrows=3):
    
    '''
    Get a list of correctly placed and rotated 
    arrows/markers to be plotted
    
    Parameters
    locations : list of lists of lat lons that represent the 
                start and end of the line. 
                eg [[41.1132, -96.1993],[41.3810, -95.8021]]
    arrow_color : default is 'blue'
    size : default is 6
    n_arrows : number of arrows to create.  default is 3
    Return
    list of arrows/markers
    '''
    
    Point = namedtuple('Point', field_names=['lat', 'lon'])
    
    # creating point from our Point named tuple
    p1 = Point(locations[0][0], locations[0][1])
    p2 = Point(locations[1][0], locations[1][1])
    
    # getting the rotation needed for our marker.  
    # Subtracting 90 to account for the marker's orientation
    # of due East(get_bearing returns North)
    rotation = get_bearing(p1, p2) - 90
    
    # get an evenly space list of lats and lons for our arrows
    # note that I'm discarding the first and last for aesthetics
    # as I'm using markers to denote the start and end
    arrow_lats = np.linspace(p1.lat, p2.lat, n_arrows + 2)[1:n_arrows+1]
    arrow_lons = np.linspace(p1.lon, p2.lon, n_arrows + 2)[1:n_arrows+1]
    
    arrows = []
    
    #creating each "arrow" and appending them to our arrows list
    for points in zip(arrow_lats, arrow_lons):
        arrows.append(folium.RegularPolygonMarker(location=points, 
                      fill_color=color, number_of_sides=3, 
                      radius=size, rotation=rotation).add_to(trip_map))
    return arrows
```

```python
def get_bearing(p1, p2):
    
    '''
    Returns compass bearing from p1 to p2
    
    Parameters
    p1 : namedtuple with lat lon
    p2 : namedtuple with lat lon
    
    Return
    compass bearing of type float
    
    Notes
    Based on https://gist.github.com/jeromer/2005586
    '''
    
    long_diff = np.radians(p2.lon - p1.lon)
    
    lat1 = np.radians(p1.lat)
    lat2 = np.radians(p2.lat)
    
    x = np.sin(long_diff) * np.cos(lat2)
    y = (np.cos(lat1) * np.sin(lat2) 
        - (np.sin(lat1) * np.cos(lat2) 
        * np.cos(long_diff)))
    bearing = np.degrees(np.arctan2(x, y))
    
    # adjusting for compass bearing
    if bearing < 0:
        return bearing + 360
    return bearing
```
